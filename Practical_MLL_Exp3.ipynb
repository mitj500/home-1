```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.jointplot(x='Temperature',y='Revenue',data=df)
```

```python
y=df['Revenue']
X=df[['Temperature']]
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(fit_intercept=True)
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_test
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Revenue Generated vs. Temperature Ice Cream Stand(Training dataset)')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
```

```python
# Below is execution of 50_startups dataset using LinearRegresssion
```

```python
ds = pd.read_csv('/content/50_Startups.csv')
x = ds.iloc[:, :-1].values
y = ds.iloc[:, -1].values
```

```python
ds.head(10)
```

```python
ds.isnull().any()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# Apply OneHotEncoder to the categorical feature at index 3
ct = ColumnTransformer(transformers=[('encoding', OneHotEncoder(), [3])], remainder='passthrough')
x = np.array(ct.fit_transform(x))

print(x)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
#
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(-1,1), y_test.reshape(-1,1)),1))
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,lr.predict(X_test))
```
